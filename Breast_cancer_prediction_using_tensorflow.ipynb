## Split the training and test data

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_dataframe, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train, y_train
```

## Creating a Neural Network

```python
import tensorflow as tf
```

```python
# Let's create model
tf.random.set_seed(42)
inputs = tf.keras.Input(shape=(30,))
x = tf.keras.layers.Dense(20, activation='relu')(inputs)
x = x = tf.keras.layers.Dense(10, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model_1 = tf.keras.Model(inputs,outputs);
model_1.summary()
```

```python
model_1.compile(loss = tf.keras.losses.BinaryCrossentropy(),
                     optimizer=tf.keras.optimizers.Adam(),
                     metrics=['accuracy'])
```

```python
history_1 = model_1.fit(X_train,y_train,validation_split=0.1, epochs=10)
```

```python
model_1_pred_probs = model_1.predict(X_test)
model_1_pred_probs[:5]
```

```python
model_1_preds = tf.squeeze(tf.round(model_1_pred_probs),axis=-1)
model_1_preds[:5]
```
